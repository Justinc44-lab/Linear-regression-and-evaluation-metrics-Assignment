Theoretical Answers**

#### **1. What does R-squared represent in a regression model?**  
R-squared measures the proportion of variance in the dependent variable explained by the independent variables. It ranges from 0 to 1, where a higher value indicates a better fit.

#### **2. What are the assumptions of linear regression?**  
A. Linearity – The relationship between independent and dependent variables should be linear.  
B. Independence – Observations should be independent.  
C. Homoscedasticity – Constant variance of residuals.  
D. Normality – Residuals should be normally distributed.  
E. No multicollinearity – Independent variables should not be highly correlated.

#### **3. What is the difference between R-squared and Adjusted R-squared?**  
A. R-squared measures the proportion of explained variance.  
B. Adjusted R-squared adjusts for the number of predictors, preventing overestimation in models with more variables.

#### **4. Why do we use Mean Squared Error (MSE)?**  
MSE measures the average squared difference between actual and predicted values, penalizing large errors and ensuring differentiability for optimization.

#### **5. What does an Adjusted R-squared value of 0.85 indicate?**  
It means that 85% of the variance in the dependent variable is explained by the independent variables, considering the number of predictors in the model.

#### **6. How do we check for normality of residuals in linear regression?**  
A. Histogram – Residuals should follow a bell curve.  
B. Q-Q Plot – Points should align along the diagonal.  
C. Shapiro-Wilk Test – Statistical test for normality.  

#### **7. What is multicollinearity, and how does it impact regression?**  
Multicollinearity occurs when independent variables are highly correlated, leading to unreliable coefficient estimates and increased variance.

#### **8. What is Mean Absolute Error (MAE)?**  
MAE is the average absolute difference between actual and predicted values, making it less sensitive to outliers than MSE.

#### **9. What are the benefits of using an ML pipeline?**  
A. Automates preprocessing, training, and evaluation.  
B. Ensures consistency and reproducibility.  
C. Reduces manual intervention.  
D. Enhances model deployment.

#### **10. Why is RMSE considered more interpretable than MSE?**  
RMSE is in the same unit as the dependent variable, making error magnitude easier to understand.

#### **11. What is pickling in Python, and how is it useful in ML?**  
Pickling serializes and saves ML models for later use, enabling model reuse without retraining.

#### **12. What does a high R-squared value mean?**  
A high R-squared value indicates that the independent variables explain most of the variance in the dependent variable.

#### **13. What happens if linear regression assumptions are violated?**  
A. Biased or inefficient estimates.  
B. Inaccurate predictions.  
C. Misleading statistical inferences.  
D. Increased model variance.

#### **14. How can we address multicollinearity in regression?**  
A. Remove highly correlated variables.  
B. Use Principal Component Analysis (PCA).  
C. Apply Ridge or Lasso regression.  
D. Compute Variance Inflation Factor (VIF) to detect multicollinearity.

#### **15. How can feature selection improve model performance in regression analysis?**  
A. Reduces overfitting.  
B. Improves interpretability.  
C. Enhances computational efficiency.  
D. Reduces noise and improves accuracy.

#### **16. How is Adjusted R-squared calculated?**  
Adjusted R² = 1 - [(1 - R²) * (n - 1) / (n - p - 1)]  
Where n = number of observations, p = number of predictors.

#### **17. Why is MSE sensitive to outliers?**  
MSE squares the errors, amplifying the impact of large deviations caused by outliers.

#### **18. What is the role of homoscedasticity in linear regression?**  
Homoscedasticity ensures that residuals have constant variance, leading to unbiased and reliable parameter estimates.

#### **19. What is Root Mean Squared Error (RMSE)?**  
RMSE is the square root of MSE, measuring the model’s average prediction error in the same unit as the target variable.

#### **20. Why is pickling considered risky?**  
A. Security risk – Can execute malicious code if tampered with.  
B. Compatibility issues – Pickled objects may not work across Python versions.  
C. Large file sizes – Inefficient storage for large models.

#### **21. What alternatives exist to pickling for saving ML models?**  
A. Joblib – Optimized for large NumPy arrays.  
B. ONNX – Cross-platform model format.  
C. PMML – Standardized XML format.  
D. TensorFlow SavedModel – For deep learning models.

#### **22. What is heteroscedasticity, and why is it a problem?**  
Heteroscedasticity occurs when residual variance changes across values of the independent variable, leading to inefficient estimators and unreliable significance tests.

#### **23. How can interaction terms enhance a regression model's predictive power?**  
A. Capture relationships between independent variables.  
B. Improve model accuracy by explaining additional variance.  
C. Reveal hidden patterns in data.  


```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 1. Write a Python script to visualize the distribution of errors (residuals) for a multiple linear regression model using Seaborn's "diamonds" dataset.
print("\n1. Visualizing Residuals")
diamonds = sns.load_dataset('diamonds')
X = diamonds[['carat', 'depth', 'table']]  # Selecting features
y = diamonds['price']
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)
residuals = y - y_pred
sns.histplot(residuals, kde=True)
plt.title("Residuals Distribution")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

# 2. Write a Python script to calculate and print Mean Squared Error (MSE), Mean Absolute Error (MAE), and Root Mean Squared Error (RMSE) for a linear regression model.
print("\n2. Error Metrics")
mse = mean_squared_error(y, y_pred)
mae = mean_absolute_error(y, y_pred)
rmse = np.sqrt(mse)
print(f"MSE: {mse}, MAE: {mae}, RMSE: {rmse}")
```

```python
# 3. Write a Python script to check if the assumptions of linear regression are met. Use a scatter plot to check linearity, residuals plot for homoscedasticity, and correlation matrix for multicollinearity.
print("\n3. Checking Assumptions")
sns.scatterplot(x=y_pred, y=residuals)
plt.axhline(0, color='r', linestyle='--')
plt.title("Residuals vs. Predictions")
plt.show()

corr_matrix = X.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Feature Correlation Matrix")
plt.show()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score

# 4. Write a Python script that creates a machine learning pipeline with feature scaling and evaluates the performance of different regression models.
def regression_pipeline(X, y):
    pipeline = Pipeline([
        ('scaler', StandardScaler()),
        ('regressor', LinearRegression())
    ])
    scores = cross_val_score(pipeline, X, y, scoring='r2', cv=5)
    print(f"Cross-validated R²: {np.mean(scores)}")

regression_pipeline(X, y)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# 11. Write a Python script that fits polynomial regression models of different degrees to a synthetic dataset and compares their performance.
X_syn = np.random.rand(100, 1) * 10
y_syn = 3 * X_syn**2 + 2 * X_syn + 5 + np.random.randn(100, 1) * 5

degrees = [1, 2, 3, 4]
for d in degrees:
    poly = PolynomialFeatures(degree=d)
    X_poly = poly.fit_transform(X_syn)
    model = LinearRegression()
    model.fit(X_poly, y_syn)
    y_pred = model.predict(X_poly)
    plt.scatter(X_syn, y_syn, label="Actual Data", alpha=0.5)
    plt.scatter(X_syn, y_pred, label=f"Degree {d} Prediction", alpha=0.5)
    plt.legend()
    plt.title(f"Polynomial Regression (Degree {d})")
    plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# 16. Write a Python script that creates a machine learning pipeline with data standardization and a multiple linear regression model, and prints the R-squared score.
X_syn = np.random.rand(100, 3) * 10
y_syn = 3 * X_syn[:, 0] + 2 * X_syn[:, 1] + 5 * X_syn[:, 2] + 10 + np.random.randn(100) * 5

X_train, X_test, y_train, y_test = train_test_split(X_syn, y_syn, test_size=0.2, random_state=42)

pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('regressor', LinearRegression())
])

pipeline.fit(X_train, y_train)
r2_score = pipeline.score(X_test, y_test)
print(f"R² Score: {r2_score}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split

# 25. Write a Python script that compares polynomial regression models of different degrees and prints the R-squared score for each.
X_syn = np.random.rand(100, 1) * 10
y_syn = 2 * X_syn**3 + 3 * X_syn**2 + 4 * X_syn + 5 + np.random.randn(100, 1) * 10

degrees = [1, 2, 3, 4, 5]
for d in degrees:
    poly = PolynomialFeatures(degree=d)
    X_poly = poly.fit_transform(X_syn)
    X_train, X_test, y_train, y_test = train_test_split(X_poly, y_syn, test_size=0.2, random_state=42)

    model = LinearRegression()
    model.fit(X_train, y_train)
    r2_score = model.score(X_test, y_test)

    print(f"Degree {d} Polynomial Regression - R² Score: {r2_score}")
```
